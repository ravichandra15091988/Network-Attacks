# Model 1 

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix
```

```python
feature_names = [x for x in train.columns if x not in ['connection_id','target']]
target = train['target']
feature_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,target,train_size=0.7,stratify=target,test_size=0.3,random_state=2017)
```

```python
clf1 = DecisionTreeClassifier(max_depth=8, min_samples_split = 7, max_features = 'sqrt')
```

```python
## Train model
clf1.fit(train[feature_names], target)
```

```python
#prediction on test
pred1=clf1.predict(test[feature_names])
```

```python
clf1_accu= cross_val_score(clf1, train[feature_names],target, cv=5, scoring='accuracy')
clf1_accu
```

```python
##make submission
sub=pd.read_csv("sample_submission.csv")
sub['target'] = pred1
sub['target'] = sub['target'].astype(int)
sub.to_csv('sub1.csv', index=False)
```
